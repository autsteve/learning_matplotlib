```python
# 导入numpy包，在程序中简写为np
import numpy as np

import matplotlib.pyplot as plt
```

## **常见图像**

### **绘制$\sin{x}(0\le x \le 10\pi)$的图像**

````python
import numpy as np
import matplotlib.pyplot as plt

X =np.linspace(0, 10*np.pi, 4000)
Y =np.sin(X)

fig,ax =plt.subplots()
ax.plot(X,Y)
plt.show()
````

### **绘制散点图(scatter)**

````python
# 因为本页面已经导入过matplotlib和numpy，不必重复导入
X = np.random.uniform(0,1,100)
Y = np.random.uniform(0,1,100)
fig,ax =plt.subplots()
ax.scatter(X, Y)
plt.show()
````

### **绘制柱状图(bar)**

````python
X = np.arange(10)
Y = np.random.uniform(1,10,10)
fig,ax =plt.subplots()
ax.bar(X,Y)
plt.show()
````

### **绘制灰度图(imshow)**

````python
Z = np.random.uniform(0,1,(8,8))
fig,ax = plt.subplots()
ax.imshow(Z)
plt.show()
````

### **绘制二维等高线图(contourf)**

````python
Z = np.random.uniform(0,1,(8,8))
fig, ax = plt.subplots()
ax.contourf(Z)
plt.show()
````

### **绘制饼图(pie)**

````python
Z = np.random.uniform(0,1,4)
fig,ax = plt.subplots()
ax.pie(Z)
plt.show()
````

### **绘制直方图(hist)**

````python
Z = np.random.uniform(0,1,100)
fig,ax = plt.subplots()
ax.hist(Z)
plt.show()
````

### **绘制含误差条的线图(errorbar)** 

````python
X = np.arange(5)
Y = np.random.uniform(0,1,5)

fig,ax = plt.subplots()

ax.errorbar(X,Y,Y/4)
plt.show()
````

### **绘制箱线图(boxplot)**

````python
Z = np.random.normal(0,1,(100,3))
fig,ax = plt.subplots()
ax.boxplot(Z)
plt.show()
````

## **线条设置 | 以$\sin{x}(0\le x\le2\pi)$为例** 

### **设置线条颜色** 

````python
X = np.linspace(0,10,100)
Y = np.sin(X)
fig,ax =plt.subplots()
ax.plot(X,Y, color = 'red')
plt.show()
````

### **设置线型**

````python
X = np.linspace(0,10,100)
Y = np.sin(X)
fig,ax =plt.subplots()
ax.plot(X,Y, linestyle = '--')
plt.show()
````

### **设置线宽** 

````python
X = np.linspace(0,10,100)
Y = np.sin(X)
fig,ax =plt.subplots()
ax.plot(X,Y, linewidth = 5)
plt.show()
````

### **描点形状设置**

````python
X = np.linspace(0,10,100)
Y = np.sin(X)
fig,ax =plt.subplots()
ax.plot(X,Y, marker = 'o')
plt.show()
````

## **图像排布设置**

### **多个函数图像共用一个坐标系**

````python
X = np.linspace(-2*np.pi,2*np.pi,100)
Y1 = np.sin(X)
Y2 = np.cos(X)
fig,ax = plt.subplots() #图像1行1列
ax.plot(X,Y1,X,Y2)      
plt.show()
````

### **图像上下排布**

````python
fig, (ax1,ax2) = plt.subplots(2,1) # 图像2行1列
ax1.plot(X,Y1,color ='C0')
ax2.plot(X,Y2,color ='C1')
plt.show()
````

### **图像左右排布**

````python
fig, (ax1,ax2) = plt.subplots(1,2) #图像1行2列
ax1.plot(Y1,X,color ='C0')
ax2.plot(Y2,X,color ='C1')
plt.show()
````

## **标题、标签设置** 

### **标题设置**

````python
X = np.linspace(-np.pi,np.pi,100)
Y = np.sin(X)
fig,ax = plt.subplots()
ax.plot(X,Y)
ax.set_title('A Sine wave')
plt.show()
````

### **标签设置** 

````python
X = np.linspace(-np.pi,np.pi,100)
Y = np.sin(X)
fig,ax = plt.subplots()
ax.plot(X,Y)
ax.set_xlabel('Time') #横坐标设置
ax.set_ylabel('wave') #纵坐标设置
plt.show()
````
